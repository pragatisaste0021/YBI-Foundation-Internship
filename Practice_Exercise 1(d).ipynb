##**Import library**

```python
import pandas as pd
```

##**Insert Tips dataset and store as the pandas dataframe with name tipspayment**

```python
tipspayment = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Tips%20Payment%20Data.csv')
```

##**Display the first 5 rows of tipspayment dataframe**

```python
tipspayment.head()
```

##**Calculate percentage of tip to total bill**

```python
tipspayment['Tip']/tipspayment['Total Bill']*100
```

##**Create a new column of percentage tip**

```python
tip_percentage = tipspayment['Tip']/tipspayment['Total Bill']*100
```

```python
tip_percentage
```

##**Insert percentage tip in existing tipspayment dataframe**

```python
tipspayment['tip_percentage'] = tipspayment['Tip']/tipspayment['Total Bill']*100
```

```python
tipspayment.head()
```

##**Round upto one decimal place the tip_percentage column values**

```python
tipspayment['tip_percentage'] = tipspayment['tip_percentage'].round(1)
```

```python
tipspayment.head()
```

##**Drop column Time**

```python
tipspayment = tipspayment.drop(['Time'],axis=1)
```

```python
tipspayment.head()
```

##**Index tipspayment dataframe as per Payment ID**

```python
tipspayment.set_index('Payment ID')
```

```python
tipspayment.head()
```

##**Change index tipspayment dataframe as per Payment ID**

```python
tipspayment = tipspayment.set_index('Payment ID')
```

```python
tipspayment.head()
```

##**Reset index of tipspayment dataframe to row index**

```python
tipspayment = tipspayment.reset_index()
```
