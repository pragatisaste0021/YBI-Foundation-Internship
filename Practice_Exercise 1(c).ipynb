##**Select 100th row and all columns with iloc function**

##**Select all rows with columns label name and fare column with iloc function**

```python
titanic.iloc[:,[2,8]]
```

##**Select all rows with loc function and column label name and fare**

```python
titanic.loc[:,['name','fare']]
```

##**Select row number 50th, 25th, 15th and column label passenger class, fare, age with both loc and iloc function**

```python
titanic.loc[[50,25,15],['pclass','fare','age']]
```

```python
titanic.iloc[[50,25,15],[0,8,4]]
```

##**Select rows from 10th, 25th and column label passenger class, fare, age with both loc and iloc function**

```python
titanic.loc[10:25,['pclass','fare','age']]
```

```python
titanic.iloc[10:26,[0,8,4]]
```

##**Select rows from 10th to 15th and columns from passenger class to age with both loc and iloc function**

```python
titanic.loc[10:15,'pclass':'age']
```

```python
titanic.iloc[10:16,0:5]
```

##**Select all passengers with age equal to and more than 35 years**

```python
titanic[titanic['age']>=35]
```

##**Select all passengers with age equal to and more than 35 years and column with label passenger class to age**

```python
titanic.loc[(titanic['age']>=35),'pclass':'age']
```

##**Select all female passengers with age equal to and more than 35 years**

```python
titanic.loc[(titanic['age']>=35) & (titanic['sex']=='female')]
```
